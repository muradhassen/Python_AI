```python
#count data in dataFrame
import pandas as pd
import numpy as np
from numpy.random import randn
from google.colab import files
df = pd.read_csv('ai_job_dataset.csv')
df['company_name'].value_counts()  #count distinct value -desindeing
df['company_name'].value_counts(ascending=True)  #count distinct value -aceending
df['company_name'].value_counts(dropna=False, ascending=True) #show null value aslo NaN
df['education_required'].value_counts(dropna=False, ascending=True , normalize=True) #show percentage
df['education_required'].value_counts()['PhD']
df['education_required'].value_counts()
df.groupby('education_required').count()
df.groupby('education_required').size()
# df.apply(pd.value_counts) #get a count of all colums across all colums
```
